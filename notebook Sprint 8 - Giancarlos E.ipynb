- Paso 5. Prueba de hipótesis (Python)

<div class="alert alert-block alert-warning"> 
<b>Comentario del Revisor #1</b> <a class="tocSkip"></a>

Es necesario volver a importar la librería _pandas_?

</div>

```python
import pandas as pd
from scipy import stats

# Cargar datos
df = pd.read_csv('/datasets/project_sql_result_07.csv')

# Separar grupos según condiciones climáticas
good = df[df['weather_conditions'] == 'Good']['duration_seconds']
bad = df[df['weather_conditions'] == 'Bad']['duration_seconds']

alpha = 0.05

stat, p_value1 = stats.ttest_ind(good, bad, equal_var=False)
print("Estadístico t:", stat)
print("p-value:", p_value1)
if p_value1 < alpha:
    print("Conclusión: Rechazamos Hipotesis Nula → Las duraciones promedio son diferentes.\n")
else:
    print("Conclusión: No rechazamos Hipotesis Nula → Las duraciones promedio son iguales.\n")
```

<div class="alert alert-block alert-warning"> 
<b>Comentario del Revisor #1</b> <a class="tocSkip"></a>

Al realizar una prueba de hipótesis, es muy importante verificar los supuestos. El t-test que usaste funciona bien cuando se cumplen ciertos supuestos, como que las varianzas de ambos grupos (sábados lluviosos y no lluviosos) sean similares. Puedes realizar la prueba de levene para comprobar que los dos grupos tienen la misma varianza y, de esta forma, podrás saber el valor del parámetro equal_var que se debe utilizar en la función stats.ttest_ind.

</div>

```python
#Grafico Boxplot GOOD Vs Bad

df['duration_seconds'] = pd.to_numeric(df['duration_seconds'], errors='coerce')

plt.figure(figsize=(8,6))
df.boxplot(column='duration_seconds', by='weather_conditions')
plt.title('Duración de viajes Loop → O’Hare por condiciones climáticas (sábados)')
plt.suptitle('')
plt.xlabel('Condiciones climáticas')
plt.ylabel('Duración del viaje (segundos)')
plt.show()
```

```python
#grafico de barras GOOD Vs BAD
group_means = df.groupby('weather_conditions')['duration_seconds'].mean()

plt.figure(figsize=(6,4))
plt.bar(group_means.index, group_means.values)
plt.title('Duración promedio de viajes por condiciones climáticas')
plt.xlabel('Condiciones climáticas')
plt.ylabel('Duración promedio (segundos)')
plt.show()

print(group_means)
```

- Se aplicó una prueba t de Welch,y utilizando un nivel de significación de alpha 0.05, se evaluó si la duración promedio de los viajes desde el Loop hasta el Aeropuerto O’Hare cambia los sábados lluviosos.

- La comparación entre los viajes en condiciones climáticas “Good” y “Bad” mostró que el valor p obtenido fue 6.738994326108734e-12.

- Dado que p < 0.05, rechazamos la hipótesis nula.

- Por lo tanto, sí existe evidencia estadísticamente significativa de que la duración promedio de los viajes cambia los sábados lluviosos.
